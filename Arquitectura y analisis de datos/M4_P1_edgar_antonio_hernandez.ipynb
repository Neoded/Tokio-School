Creamos la sesion se Spark, ya que estaremos utilizando python, utilizaremos la libreria pyspark, se recomienda ejecutar con un entorno virtual para instalar la libreria con pip install pyspark.

Si lo ejecutamos de forma local debemos tener instalado JAVA.

Para usar pyspark en windows debe utilizarse la version 3.11 de python.

```python
#Importar las librerias
from pyspark.sql import SparkSession
import os
import sys

os.environ["PYSPARK_PYTHON"] = sys.executable
os.environ["PYSPARK_DRIVER_PYTHON"] = sys.executable

# Crear sesión Spark
spark = SparkSession.builder \
    .appName("PracticaRDD") \
    .master("local[1]") \
    .getOrCreate()
```

Debemos obtener el sparkcontext, permite conectar python con spark, administrar los recursos, coordina la ejecucion

```python
# Obtener SparkContext
sc = spark.sparkContext
```

El ejercicio pide crear un rdd con los valores [74,22, 6,0,22, 10, 5, 10, 12]

```python
#Crear una lista en python con los valores del ejercicio
datos= [74, 22, 6, 0, 22, 10, 5, 10, 12]

#Se utiliza la funcion parallelize para convertir la lista en un rdd
rdd = sc.parallelize(datos)
```

Mostrar los valores máximos y mínimos del RDD.

Para ello utilizaremos los metodos .min() para mostrar el minimo y .max() para mostrar el maximo, estos metodos son parte del objeto rdd que creamos anteriormente

```python
#Mostrar el valor maximo de rdd
print(f'Valor maximo del rdd:{rdd.max()}')

#Mostrar el valor minimo del rdd
print(f'Valor minimo del rdd:{rdd.min()}')
```

Calcular la suma de todos los valores.

Igualmente utilizaremos un metodo del objeto rdd, .sum() para realizar la suma

```python
#Realizar la suma de todos los valores del rdd con .sum()
print(f'La suma de los elementos del rdd es: {rdd.sum()}')
```

Generar un nuevo RDD fruto de sumar diez unidades a cada elemento.

Para ello utilizaremos el metodo map, el cual recibe una funcion que aplicara a todos los elementos, utilizaremos lambda para generar una funcion anonima que sume 10 a los elementos

```python
#Sumar 10 a los elementos con map generando un nuevo rdd
nuevo_rdd = rdd.map(lambda x: x+10)

#Mostrar los resultados
nuevo_rdd.collect()
```

Generar un nuevo RDD que contenga los valores únicos del RDD original.

Para ello tambien nos apoyaremos en otro metodo, distinct().

```python
#Obtener los elementos unicos del primer rdd con distinct 
rdd_unicos = rdd.distinct()

#Mostrar los resultados
rdd_unicos.collect()
```
